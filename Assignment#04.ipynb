#### Question No.04
##### Create a dictionary where:
######     key = student name
######      value = list of marks

##### Calculate and print the percentage for each student.

```python
students_info ={
    "Ali": [85, 90, 78, 92, 88],
    "Fatima": [95, 92, 89, 94, 91],
    "Hassan": [75, 80, 82, 78, 85],
    "Abdullah": [88, 86, 90, 87, 89]
}
print("Students Average Marks")
print("Ali: ",sum(students_info["Ali"])/500*100,"%")
print("Fatima: ",sum(students_info["Fatima"])/500*100,"%")
print("Hassan: ",sum(students_info["Hassan"])/500*100,"%")
print("Abdullah: ",sum(students_info["Abdullah"])/500*100,"%")
```

#### Question No.04
##### Create a dictionary of . Use to perform the following tasks:
- Display all keys using keys()
- Display all values using values()
- Display all key-value pairs using items()
- Update a student’s score using update()
- Remove one student using pop()

```python
students_info ={
    "Ali": [85, 90, 78, 92, 88],
    "Fatima": [95, 92, 89, 94, 91],
    "Hassan": [75, 80, 82, 78, 85],
    "Abdullah": [88, 86, 90, 87, 89]
}
print("-Display all keys using keys()")
print(students_info.keys())
print("-Display all keys using Values()")
print(students_info.values())
print("-Display all key-value pairs using items()")
print(students_info.items())
print("-Update a student’s score using update(),Fatima 89 to 99")
students_info.update({"Fatima": [95, 92, 99, 94, 91]})
print(students_info.get("Fatima"))
print("-Remove one student using pop()")
students_info.pop("Hassan")
print(students_info)
```
