#### Q. 8) Write a function that takes one number and prints whether it is even or odd.

```python
def CheckEvenOdd(num):
    if num % 2 == 0:
        print(num,": your num Is Even")
    else :
        print(num,": your num Is Odd")  


CheckEvenOdd(9)
CheckEvenOdd(98)
```
